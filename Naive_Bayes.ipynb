```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
### Independent and dependent feature
X = df.iloc[:, 1:]
y = df['total_bill']
```

```python
X
```

```python
### Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
### Label and OneHot encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

le1.fit(pd.concat([X_train['sex'], X_test['sex']]))
X_train['sex'] = le1.transform(X_train['sex'])
X_test['sex'] = le1.transform(X_test['sex'])

le2.fit(pd.concat([X_train['smoker'], X_test['smoker']]))
X_train['smoker'] = le2.transform(X_train['smoker'])
X_test['smoker'] = le2.transform(X_test['smoker'])

le3.fit(pd.concat([X_train['time'], X_test['time']]))  # Nếu 'time' là cột trong cả hai bộ dữ liệu
X_train['time'] = le3.transform(X_train['time'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'), [3])], remainder='passthrough')
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB

y_train_class = np.digitize(y_train, bins=np.linspace(min(y_train), max(y_train), num=5))

# Mã hóa lại nhãn
le = LabelEncoder()
y_train_class = le.fit_transform(y_train_class)

# Khởi tạo mô hình Naive Bayes
classifier = GaussianNB()

# Huấn luyện mô hình với các nhãn phân loại
classifier.fit(X_train, y_train_class)
```

```python
y_pred = classifier.predict(X_test)
```
